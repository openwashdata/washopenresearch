```python
import pandas as pd
import numpy as np
```

```python
data_df = pd.read_excel('/Users/ydubey/Documents/GitHub/washopenresearch/journalwash4d.xlsx')
```

```python
data_df.head()
```

```python
data_df = data_df.groupby("title").agg(
    {'keywords': lambda x: ', '.join(str(i) for i in x if pd.notna(i)),
    'authors': lambda x: ', '.join(str(i) for i in x if pd.notna(i)),
    'supp_data-href': lambda x: ', '.join(str(i) for i in x if pd.notna(i)),
    'date': 'first',
    'abstract': 'first',
    'doi' : 'first',
    'article_link-href' : 'first',
    'web-scraper-start-url' : 'first'
}).reset_index()

data_df.loc[:,'volume'] = data_df.loc[:,'web-scraper-start-url'].str.extract(r'issue/(\d+)/(\d+)')[0]
data_df.loc[:,'issue'] = data_df.loc[:,'web-scraper-start-url'].str.extract(r'issue/(\d+)/(\d+)')[1]

data_df.loc[:,'date'] = pd.to_datetime(data_df.loc[:,'date'], format="%B %d %Y")

data_df.rename(columns={'article_link-href':'article_link', 'supp_data-href':'supp_data', 'web-scraper-start-url':'parent_url'}, inplace=True)

data_df.head()
```

```python
data_df.to_excel('/Users/ydubey/Documents/GitHub/washopenresearch/journalwash4d_cleaned.xlsx', index=False)
```
